### Question: Let’s say you want to calculate how many sales occurred within the first 30 minutes after the farmer’s market opened, how would you dynamically determine what  cutoff time to use?

```python
pd.read_sql_query("""

select 
    market_start_datetime,
    date_add(market_start_datetime, interval 30 minute) as plus30min,
    date_add(market_start_datetime, interval -30 minute) as 'plus-30min',
    date_sub(market_start_datetime, interval 30 minute) as sub30min
    
from
    datetime_demo
    
;

""",conn).head(10)
```

### Question: Let’s say we wanted to get a profile of each farmer’s market customer’s habits over time.
1. First purchase - date
2. Last purchase - date
3. Count of distinct purchases 

```python
pd.read_sql_query("""

select 
    customer_id,
    min(market_date) as first_pruchase_date,
    max(market_date) as last_purchase_date,
    datediff(max(market_date),min(market_date)) as active_days,
    datediff(now(),min(market_date)) as from_today,
    count(distinct(market_date)) as num_purchases
from
    customer_purchases
group by 
    customer_id
    
;

""",conn)
```

### Question: today’s date is May 31, 2019, and the marketing director of the farmer’s market wants to give infrequent customers an incentive to return to the market in June.


```python
pd.read_sql_query("""

select 
customer_id,
market_date,
LAG(market_date,1) OVER (PARTITION BY
                            customer_id
                        order by market_date) as last_purcahase,
DATEDIFF(market_date,   (Lag(market_date,1) over (partition by
                            customer_id
                        order by market_date))) as count_btwn_purchases

from
customer_purchases


;

""",conn)
```

```python
## FURTHER CONDITIONS  ON ABOVE QUERY 
```

```python
pd.read_sql_query("""

SELECT
*,
count(distinct(X.market_date)) as number_of_purcahses_made
FROM
(

select 
customer_id,
market_date,
LAG(market_date,1) OVER (PARTITION BY
                            customer_id
                        order by market_date) as last_purcahase,
DATEDIFF(market_date,   (Lag(market_date,1) over (partition by
                            customer_id
                        order by market_date))) as count_btwn_purchases

from
customer_purchases
)
AS X

WHERE 
    datediff("2019-05-31",X.market_date) between 0 and 31
GROUP BY X.customer_id
HAVING count(distinct(X.market_date)) = 1
;

""",conn)
```

```python
pd.read_sql_query("""

SELECT
    distinct
    customer_id,
    count(distinct(market_date))
FROM
    customer_purchases 
WHERE
    datediff("2019-05-31",market_date) between 0 and 31
GROUP BY customer_id
HAVING count(distinct(market_date)) = 1

;

""",conn)
```

#### Problem Statement:

### Display the employee's details like employee id, first_name, last_name, and calculate the years the employees have been working in the company as 'Total_years' till 8th June 2022 of those employees who have been working more than 28 years.

Return the columns 'employee_id', 'first_name', 'last_name', 'Total_years'.

```python
pd.read_sql_query("""

SELECT
    employee_id,
    first_name,
    last_name,
    hire_date,
    DATEDIFF("2022-06-08",hire_date)/365 as Total_years_,
    YEAR("2022-06-08")-YEAR(hire_date) as Total_years
    
FROM
    employees
WHERE
    (DATEDIFF("2022-06-08",hire_date)/365) >= 28

;

""",conn)
```

#### Problem Statement:

### Extract the day, month, year from the hire_date of the employees and save the columns as 'Day', 'Month', 'Year'. Display the the extracted columns and the details of those employees who were hired in the year 2000 and in January month and also salary is greater than 5000.

Return the columns 'employee_id', 'first_name', 'last_name', 'salary', 'hire_date', 'Day', 'Month', 'Year'.

```python
pd.read_sql_query("""

SELECT
    employee_id,
    first_name,
    last_name,
    salary,
    hire_date,
    day(hire_date) as Day,
    MONTH(hire_date) as Month,
    YEAR(hire_date) as Year
    
    
FROM
    employees
WHERE
    Year(hire_date) = 2000
    and
    MONTH(hire_date)  = 1
    and
    salary >= 5000 
    
    

;

""",conn)
```

#### Problem Statement:

### Display the manager details like first_name, last_name, manager_id, salary, department_name, and no.of years of experience as 'Experience' till 8th June 2022 of those managers who have experience as a manager is more than 25 years.

Return the columns 'first_name', 'last_name', 'manager_id', 'salary', 'department_name', 'Experience'.

```python
pd.read_sql_query("""

SELECT
        
    distinct
    m.first_name,
    m.last_name,
    e.manager_id,
    m.salary,
    (datediff("2022-06-08",m.hire_date))/365 as experience 
    
FROM
    employees as m
    join employees as e
        on e.manager_id = m.employee_id
WHERE
    m.manager_id is not null and 
    (datediff("2022-06-08",m.hire_date))/365 >= 25
ORDER BY 
    experience desc
    

;

""",conn)
```

```python
pd.read_sql_query("""

SELECT
e.first_name,
e.last_name,
e.manager_id,
e.salary,
d.department_name,
(datediff("2022-06-08",e.hire_date))/365 as Experience     
    
FROM
    employees as e
    join departments as d on d.department_id = e.department_id
WHERE 
    e.employee_id in (
                    select
                        distinct(manager_id)
                        from
                        employees
                        )
                    and
                    e.manager_id is not null 
                    
                    and
                    (datediff("2022-06-08",e.hire_date))/365  > 25

    
        
;

""",conn)
```

#### Problem Statement:

### Display the year from the hire_date as 'Year' and count the number of employees as 'Employees' joined in that year and order by the count of the number of employees in descending order using the employees table.

Return the columns 'Year', 'Employees'.

```python
pd.read_sql_query("""

SELECT
Year(hire_date) as Year,
count(*) as Employees 

from
employees
group by year(hire_date)
order by 2 desc
;

""",conn)
```

### Problem Statement:

### Display the employee details like employee_id,first_name, last_name,salary,hire_date,department_id, of those employees who were hired between the six months from the given date '1998-01-01' (use date_sub()) and also whose salary is highest in that each department.

Return the columns 'employee_id', 'first_name', 'last_name', 'salary', 'hire_date', 'department_id'.

```python
pd.read_sql_query("""

select
    x.employee_id,
    x.first_name,
    x.last_name,
    x.salary,
    x.hire_date,
    x.department_id
from
        (
        select 
            employee_id,
            first_name,
            last_name,
            salary,
            max(salary) over (partition by department_id) as max_salary,
            hire_date,
            department_id
        from
            employees
        where
            hire_date between  "1998-01-01" and date_add("1998-01-01", interval 6 month) 
        ) as x
where x.salary = x.max_salary



;

""",conn)
```

#### Problem Statement:

### Display employee name(first name and last name) as 'full_name', job title for the jobs who worked less than twelve months.

Return the columns 'full_name', 'job_title'.

```python


pd.read_sql_query("""

select 
concat(first_name,' ',last_name) 'full_name', 
job_title 

from employees emp 
    join job_history  jh  on  jh.employee_id = emp.employee_id 
    join jobs job on  jh.job_id = job.job_id  

where 
    (datediff(end_date,start_date) /30) < 12;

""",conn)
```
